```python
##Librerias 
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy as sp
```

```python
##Exportación de los datos 
darray = np.genfromtxt('datos-carro-1d.csv',skip_header= 5,delimiter=",",missing_values=np.nan)
darray
```

```python
## Tiempo y posiciones experimentales 
texp = darray[:,0]
xexp = darray[:,1]
yexp = darray[:,2]
```

```python
##Plot de los datos 
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
plt.title('Posición contra tiempo')
plt.plot(texp,xexp,marker='o',color='darkred',label=r'$x=x\left(t\right)$')
plt.ylabel(r'$x\left(t\right)$')
plt.xlabel(r'$t$')
plt.legend(fontsize=10)

plt.subplot(1,2,2)
plt.title('Posición contra tiempo')
plt.plot(texp,xexp,marker='o',color='darkred',label=r'$x=x\left(t\right)$')
plt.ylabel(r'$x\left(t\right)$')
plt.xlabel(r'$t$')
plt.xlim(1,1.5)
plt.legend(fontsize=10)
```

```python
##Ciclo para encontrar el tiempo inicial y la fila correspondiente en el array
for i in range(len(texp)):
    if  texp[i]>1.1 and texp[i]<1.2:
        t0 = texp[i]  #Tiempo inicial
        n = i         #Fila en la que se encuentran los datos a partir del tiempo inicial
        
        break
    else: 
        pass
print(f'{t0=},{n=}') 
```

```python
#Lista de tiempo y posiciones a partir del tiempo t=0
texp2 = texp[n:] 
xexp2 = xexp[n:]
yexp2 = yexp[n:]
```

```python
##Funcion de derivada central 
def central(X,t):
    '''Se aplicara el algoritmo de derivada central salvo en los extremos, 
    donde se realizara una derivada backward y una forward'''
    V = np.zeros(len(t))
    h = np.abs((t[-1]-t[0])/len(t))
    for i in range(len(t)):
        if i == 0:
            V[i] = (X[i+1]-X[i])/h
        elif i == len(t)-1:
            V[i] = (X[i]-X[i-1])/h
        else:
            V[i] = (X[i+1]-X[i-1])/(2*h)
    return V    

V = central(xexp2,texp2)
V
```

```python
##Ajuste lineal

#Funcion de ajuste
def velocity(t,v0,a):
    return v0 + a*t

#Ajuste
coe, Ecoe = sp.optimize.curve_fit(velocity,texp2,V)
v0 = coe[0]
a  = coe[1]
#Velocidades obtenidas por medio del ajuste
Vt = v0 + a*texp2

#Calculo de errores
Error = np.sqrt(np.diag(Ecoe))
v0_error, a_error  = Error
```

```python
##Fit con todos los datos
plt.figure(figsize=(10,8))
plt.title(r'Velocidad contra tiempo',fontsize=10)
plt.plot(texp2,Vt,color='darkgreen',label=f'Ajuste lineal \n $v_0=${v0:.3f},$a=${a:.3f},\n Error $v_0=${v0_error:.3f},Error $a=${a_error:.3f}',linewidth=2.8)
plt.scatter(texp2,V,color='darkred',s=40,label=r'Velocidad numérica')
plt.scatter(texp2,darray[n:,3],s=45,color='darkblue',marker='*',label=r'Velocidad Experimental')
plt.ylabel(r'$V\left(t\right)$')
plt.xlabel(r'$t$')
plt.legend(fontsize=10)
plt.show()
```

```python
##Opción 1 para comparar 

#Regresión no lineal de los datos 
def position(t,x0,vx0,ax):
    return x0 +vx0*t + ax*(t**2)*(1/2)
    
coe2, Ecoe2 = sp.optimize.curve_fit(position,texp2,xexp2)


#Datos a partir de la regresión no lineal
xt  = coe2[0] + coe2[1]*texp2 + coe2[2]*(texp2**2)*(1/2)
vxt = coe2[1] + coe2[2]*texp2
axt = coe2[2]

#Error regresión no lineal
Error2 = np.sqrt(np.diag(Ecoe2))
x0_error, v0_error2, a_error2  = Error2

print(f'aceleración regresión no lineal = {axt:.3f}, con error igual a {a_error2:.3f}')
```

```python
##Opción 2 para comporar 

#Error porcentual entre la aceleración promedio y la celeración de la regresión
Promedio = np.mean(darray[n:-2,5])
Error3    = np.sqrt(np.sum((darray[n:-2,5]-Promedio)**2))
print(f'aceleración Promedio = {Promedio:.3f}, Error={Error3:.3f}')
```

```python
plt.figure(figsize=(10,8))
plt.title(r'Aceleración contra tiempo',fontsize=10)
plt.scatter(texp2,darray[n:,5],color='darkred',label='Aceleración experimental',s=40)
plt.plot(texp2,np.full(len(texp2),a),color='darkblue',label=f'aceleración regresión lineal = {a:.3f}\n Error={a_error:.3f}')
plt.plot(texp2,np.full(len(texp2),axt),color='darkgreen',label=f'aceleración regresión no lineal = {axt:.3f}\n Error={a_error2:.3f}')
plt.plot(texp2,np.full(len(texp2),Promedio),color='purple',label=f'aceleración promedio = {Promedio:.3f}\n Error={Error3:.3f}')
plt.ylabel(r'$a\left(t\right)$')
plt.xlabel(r'$t$')
plt.legend(fontsize=10)
plt.show()
```
