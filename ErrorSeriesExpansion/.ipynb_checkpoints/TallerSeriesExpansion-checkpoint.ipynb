```python
#Daniel Felipe Rojas Paternina
#Jorge Luis Jaramillo Gaitán
#Federico Villadiego Forero
#Jhon sebastian palacio correa
#Diego Alejandro Campos Mendez
import numpy as np
import math 
import matplotlib.pyplot as plt
%matplotlib inline


def exp_expansion_new(x, N, fname):
    """Add docs here"""
    total = 0.0
    a_i = 1.0
    total2 = 0.0
    fout = open(fname, "w")
    for i in range(N+1):
        try:
            total  += (-1)**i*(x**i/factorial(i))
            total2 += a_i
            dif1    = np.abs(1.0-total/np.exp(-x))
            dif2    = np.abs(1.0-total2/np.exp(-x))
            a_i = -(x*a_i)/(i+1)
            fout.write(f"{i},{total},{total2},{dif1},{dif2}\n")
        except "Numerical result out of range":
            break

    
def factorial(n):
    """Add docs here"""
    result = 1.0
    for ii in range(1,n+1):
        result *= ii
    return result
 
# Call the NEW function  
exp_expansion_new(x=1.8766, N=20, fname="sumdatanew.txt")
```

```python
%matplotlib notebook
# Plot the data

import matplotlib.pyplot as plt
import numpy as np
x, y, z, k, p = np.loadtxt('sumdatanew.txt',delimiter=',' , unpack = True)

plt.subplot(1,2,1)
plt.plot(x, y, 'ro')
plt.plot(x, z, 'b-')
plt.xlabel(r'paso')
plt.ylabel(r'$\exp(-1.8766)$')

plt.subplot(1,2,2)
plt.plot(x, k, 'ro')
plt.plot(x, p, 'b-')
plt.xlabel(r'paso')
plt.ylabel(r'Diferencia relativa')
```
