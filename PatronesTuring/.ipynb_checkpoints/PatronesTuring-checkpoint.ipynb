**I.** Solución de las ecuaciones. 


   


```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt

#Constantes y parametros de la simulación
a     = 0.00028
b     = 0.005
tau   = 0.1
kappa = -0.005

dt  = 0.001
h   = 0.02
```

```python
T = np.arange(0,10+dt,dt)
X = np.arange(0,1+h,h)
Y = np.arange(0,1+h,h)
```

```python
#Condiciones iniciales aleatorias
Phi = np.zeros([len(T),len(X),len(Y)])
Psi = np.zeros([len(T),len(X),len(Y)])

for i in range(len(X)):
    for j in range(len(Y)):
        Psi[0,i,j] = np.random.random()
        Phi[0,i,j] = np.random.random()

plt.subplot(1,2,1)
plt.imshow(Psi[0],extent=(min(X),max(X),min(Y),max(Y)),origin='lower',cmap='magma',interpolation='bilinear')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.colorbar()
plt.subplot(1,2,2)
plt.imshow(Phi[0],extent=(min(X),max(X),min(Y),max(Y)),origin='lower',cmap='magma',interpolation='bilinear')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.colorbar()


plt.show()        
```

```python
#condiciones de frontera
for i in range(len(Y)):
    Phi[1:,0,:]  = Phi[1:,i,:]
    Phi[1:,-1,:] = Phi[1:,-1-i,:]
    
    Psi[1:,0,:]  = Psi[1:,i,:]
    Psi[1:,-1,:] = Psi[1:,-1-i,:]
    
for j in range(len(X)):
    Phi[1:,:,0]  = Phi[1:,:,j]
    Phi[1:,:,-1] = Phi[1:,:,-1-j]
    
    Psi[1:,:,0]  = Psi[1:,:,j]
    Psi[1:,:,-1] = Psi[1:,:,-1-j]    
    
```

```python
#Plot de condiciones de frontera
plt.subplot(1,2,1)
plt.imshow(Psi[1],extent=(min(X),max(X),min(Y),max(Y)),origin='lower',cmap='magma',interpolation='bilinear')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.colorbar()
plt.subplot(1,2,2)
plt.imshow(Phi[1],extent=(min(X),max(X),min(Y),max(Y)),origin='lower',cmap='magma',interpolation='bilinear')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.colorbar()


plt.show()  
```

```python
#Solución de la ecuación diferencial
for t in range(len(T)-1):
    for i in range(1,len(X)-1):
        for j in range(1,len(Y)-1):
            #Laplacianos
            LapPhi = (Phi[t,i+1,j] + Phi[t,i-1,j] + Phi[t,i,j+1] + Phi[t,i,j-1] - 4* Phi[t,i,j])/h**2
            LapPsi = (Psi[t,i+1,j] + Psi[t,i-1,j] + Psi[t,i,j+1] + Psi[t,i,j-1] - 4* Psi[t,i,j])/h**2
            #Soluciones
            Phi[t+1,i,j] = Phi[t,i,j] + dt*(a*LapPhi + Phi[t,i,j] - Phi[t,i,j]**3 - Psi[t,i,j] + kappa)
            Psi[t+1,i,j] = Psi[t,i,j] + dt*(b*LapPsi + Phi[t,i,j] - Psi[t,i,j])/tau
            
```
