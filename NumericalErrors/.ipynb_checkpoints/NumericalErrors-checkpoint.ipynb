## Ecuación cuadratica

```python
#Librerias 
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```python
#Función general
def cuadratic(a=1.0,b=1.0,n=10):
    """ a: Coeficiente del término cuadrático, por defecto, es igual a 1
        b: Coeficiente del término de grado 1, por defecto, es igual a 1
        n: rango máximo del exponente para el término de grado cero, por defecto, igual a 10
        
        La función retorna dos arrey. El primero corresponde a un array de nx4, 
        cuyos componentes son las raíces de la ecuación cuadrática calculadas para 
        los n distintos valores para c. El segundo array  de nx2 corresponde a los 
        errores relativos entre las raíces calculadas por medio de las dos distintas 
        fórmulas"""
    
    Solution = np.zeros([n,4])
    Errors   = np.zeros([n,2])
    
    for i in range(1,n+1):
        c  = 10**(-i)
        
        x1 = (-b+np.sqrt(b**2-4*a*c))/(2*a)
        x2 = (-b-np.sqrt(b**2-4*a*c))/(2*a)
        y1 = (-2*c)/(b+np.sqrt(b**2-4*a*c))
        y2 = (-2*c)/(b-np.sqrt(b**2-4*a*c))
        
        Solution[i-1,0] = x1
        Solution[i-1,1] = x2
        Solution[i-1,2] = y1
        Solution[i-1,3] = y2
        
        Errors[i-1,0] = np.abs((x1-y1)/x1)
        Errors[i-1,1] = np.abs((x2-y2)/x2)
        
    return Solution, Errors

Solution, Errors = cuadratic(1,1,10)
```

```python
#Plot de los valores para las raíces en función del valor de n 
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
plt.plot(np.linspace(1,10,10),Solution[:,0],color='red',label=r'$x_1$',marker='o')
plt.plot(np.linspace(1,10,10),Solution[:,2],color='blue',label=r"$x'_1$",marker='o')
plt.xlabel('n')
plt.title('primera raíz')
plt.legend(fontsize=10)

plt.subplot(1,2,2)
plt.plot(np.linspace(1,10,10),Solution[:,1],color='red',label=r'$x_2$',marker='o')
plt.plot(np.linspace(1,10,10),Solution[:,3],color='blue',label=r"$x'_2$",marker='o')
plt.xlabel('n')
plt.title('segunda raíz')
plt.legend(fontsize=10)
```

```python
#Plot de los errores relativos en función del valor de n
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
plt.title(r"Error relativo $x_1$ y $x'_1$")
plt.plot(np.linspace(1,10,10),Errors[:,0],marker ='o',color='red',label=r"$\frac{x_1-x'_1}{x_1}$")
plt.xlabel(r'$n$')
plt.legend(fontsize=10)

plt.subplot(1,2,2)
plt.title(r"Error relativo $x_2$ y $x'_2$")
plt.plot(np.linspace(1,10,10),Errors[:,1],marker ='o',color='blue',label=r"$\frac{x_2-x'_2}{x_2}$")
plt.xlabel(r'$n$')
plt.legend(fontsize=10)
```

Pese a que en un principio se evidencia el mismo comportamiento en los valores de las raíces en función de $n$, se puede evidenciar una diferencia al analizar los errores relativos entre $x_1$ y $x'_1$ y los errores relativos entre $x_2$ y $x'_2$ en función de n. Recordemos que el valor del coeficiente de orden cero es igual a $ c = 10^{-n}$, por lo cual en torno $n$ se hace más grande $c$ se hace más pequeño, hasta el punto en que se genera un underflows y la máquina toma el valor como nulo. Esto puede ocasionar errores de overflows y underflows y es lo que ocasiona un aumento en el error relativo a partir de $n=8$. Para la primera raíz, tendremos que para $x_1$, valores muy pequeños de $n$ hacen que el númerador se anule produciendo un underflows, por lo cual en este caso resulta mejor utilizar la segunda relación y por ende tomar el valor $x'_1$. Por lo contrario, para la segunda raiz, valores muy pequeños de $n$ ocasionarán que el denominador de $x'_2$ se anule, produciendo un overflows y por ende es mejor considerar la relación usual y tomar el valor $x_2$.

## Arreglando una expresión matemática

```python
#Definción de la función original
def f(x):
    return 1-np.sin(x)

##Cantidades necesarias
#Valor central del dominio
c = np.pi/2 
#Lista inicial para el dominio
X = np.zeros(21)
#Labels para los plots
Ticks   = [-1e-2,-1e-4,-1e-6,-1e-8,0,1e-8,1e-6,1e-4,1e-2]
Name    = [r'$\frac{\pi}{2}-10^{-2}$',r'$\frac{\pi}{2}-10^{-4}$',r'$\frac{\pi}{2}-10^{-6}$',r'$\frac{\pi}{2}-10^{-8}$',r'$\frac{\pi}{2}$',r'$\frac{\pi}{2}+10^{-8}$',r'$\frac{\pi}{2}+10^{-6}$',r'$\frac{\pi}{2}+10^{-4}$',r'$\frac{\pi}{2}+10^{-2}$']
```

```python
#Solución evaluada para variaciones centradas en pi/2
F1 = np.zeros(21) 

F1[int(len(F1)/2)] = f(c)
X[int(len(F1)/2)] = c

for n in range(int(len(F1)/2)):
    C1 = c - 10**(-(n+1))
    C2 = c + 10**(-(10-n))
    F1[n] = f(C1)
    F1[n+11] = f(C2) 
    X[n] = C1
    X[n+11] = C2

F1
```

```python
#Plot de la función vs x
plt.figure(figsize=(10,6))

plt.semilogy(X-np.pi/2,F1,marker='o',color='blue',label=r"$f(x)$")
plt.xscale('symlog',linthresh = 1e-10)
plt.xticks(Ticks,Name)
plt.xlim(-0.2,0.2)
plt.ylabel(r'$1-\sin{x}$',fontsize=12)
plt.xlabel(r'$x$',fontsize=12)

plt.legend(fontsize=10)
plt.show()
```

```python
#Función corregida
def fnew(x):
    return np.cos(x)**2 / (1+np.sin(x))
```

```python
#Nueva solución evaluda para valores centrados en pi/2
F2 = np.zeros(21)


F2[int(len(F2)/2)] = fnew(c)
for n in range(int(len(F2)/2)):
    C1 = c - 10**(-(n+1))
    C2 = c + 10**(-(10-n))
    F2[n] = fnew(C1)
    F2[n+11] = fnew(C2) 

F2
```

```python
#Plot de la nueva solución 
plt.figure(figsize=(10,6))

plt.semilogy(X-np.pi/2,F2,marker='o',color='blue',label=r"$f(x)$")
plt.xscale('symlog',linthresh = 1e-10)
plt.xlim(-0.2,0.2)
plt.xticks(Ticks,Name)
plt.ylabel(r'$1-\sin{x}$',fontsize=12)
plt.xlabel(r'$x$',fontsize=12)

plt.legend(fontsize=10)
plt.show()
```

En un principio, se estaba presentando un underflow en la función original debido a que tomamos valores muy cercanos a un punto crítico de la función, sin embargo, esto puede ser corregido cambiando la expresión original por una en donde no se presenten underflows o overflows, para eso se hace uso de relaciones trigonometricas y se divide y multplica por $1+\sin{x}$ con el fin de obtener como nueva función $f(x)$ la relación 
$$ f(x) = \frac{\cos{x}^2}{\left(1+\sin{x}\right)}  $$
la cual corrige de manera satisfactoria el error númerico como se muestra en la figura.
