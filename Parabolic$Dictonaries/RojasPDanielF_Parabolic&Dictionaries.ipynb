Para encontrar la relación de altura máxima se parte de las ecuaciones de movimiento del tiro parabólico de una partícula bajo la influencia del campo gravitacional terrestre, con lo cual encontramos
\begin{equation}
    y_{max} = \frac{v_0^2\sin\left(\theta_0\right)^2}{2g}
\end{equation}
donde $v_0$ y $\theta_0$ son la velocidad y ángulo iniciales respectivamente. 

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Funtion definition
def Ymax(C0, g = 9.807):
    ''' C0:     Dictionary of initial conditions 
        v0:     Initial velocity in m/s
        theta0: Initial angle in radians
        g:      Gravity in m/s^2
        ymax:   Maximum height in meters
    '''
    ymax = (C0["v0"]**2 * np.sin(C0["theta0"])**2)/(2*g)
    return ymax
```

```python
#Maximum height for vo=2.34 and theta0∈[0,pi/2]
C0   = {"v0":2.34,"theta0":np.linspace(0,np.pi/2,20)} #Initial conditions
ymax = Ymax(C0) #List of maximum height
```

```python
#Plot of Maximum height vs angle
plt.figure(figsize=(10,6))
plt.plot(np.linspace(0,np.pi/2,20),ymax,label=r'$y_{max}$')
plt.legend(fontsize=10)
plt.xlabel(r'$\theta$ (radians)',fontsize=12)
plt.ylabel(r'$y_{max}(\theta)$ (m)',fontsize=12)
plt.title('Maximum height vs angle in parabolic Motion',fontsize=15)
plt.savefig("Ymaxvstheta.pdf")
plt.show()
```
